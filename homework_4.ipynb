```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib inline
```

```python
def dftmatrix(N):
    '''construct DFT matrix'''
   
    # create a 1xN matrix containing indices 0 to N-1
    a = np.expand_dims(np.arange(N), 0)
    
    # take advantage of numpy broadcasting to create the matrix
    WN = np.exp(-2j*np.pi*a.T*a/N)
    
    return WN
```

###Signal A

```python
N = 64
n = np.linspace(0,N-1,N)
freq = np.linspace(0, np.pi,N/2+1)


y_1 = np.cos(n/64*8*2*np.pi)
X_1  = np.fft.rfft(y_1)

plt.subplot(2,1,1)
plt.stem(n, y_1)

plt.subplot(2,1,2)
plt.stem(freq,np.abs(X_1))

```
